# Data Cleaning & Transformation

## Can you make the salary column into integers?

```python
data.columns
```

```python
data['Salary Estimate'].unique()
```

```python
data['Salary Estimate'] = data['Salary Estimate'].replace(
    {
        '\$': '',
         'K': '',
         ' \(Glassdoor est.\)': '',
         '(Employer est.)':''
    }, regex=True).str.replace('()', '')
```

```python
data['Salary Estimate'].unique()
```

```python
data['min_salary'] = data['Salary Estimate'].apply(lambda x: x.split('-')[0]).astype(int)
data['max_salary'] = data['Salary Estimate'].apply(lambda x: x.split('-')[-1]).astype(int)
data['avg_salary'] = (data['min_salary'] + data['max_salary']) / 2
data[['Salary Estimate' ,'min_salary', 'max_salary', 'avg_salary']]
```

## What information can you extract out of job descriptions?  

```python
data['Job Description']
```

* As you see there is whole bunch of information in 'Job Description' column, we can not read all these one by one. these very popular technologies used in Data Science:
    + Python
    + Excel
    + Hadoop
    + Spark
    + AWS
    + Tableau
    + Big Data
* If employer requires some of these technologies these technologies must be written in 'Job Description' column, we will create 7 boolean (0/1) columns which refer to these 7 technologies.

```python
data['python'] = data['Job Description'].map(lambda x: 1 if 'python' in x.lower() else 0)
data['excel']  = data['Job Description'].map(lambda x: 1 if 'excel' in x.lower() else 0)
data['hadoop'] = data['Job Description'].map(lambda x: 1 if 'hadoop' in x.lower() else 0)
data['spark']  = data['Job Description'].map(lambda x: 1 if 'spark' in x.lower() else 0)
data['aws'] = data['Job Description'].map(lambda x: 1 if 'aws' in x.lower() else 0)
data['tableau'] = data['Job Description'].map(lambda x: 1 if 'tableau' in x.lower() else 0)
data['big_data'] = data['Job Description'].map(lambda x: 1 if 'big data' in x.lower() else 0)
```

```python
data[['python', 'excel', 'hadoop', 'spark', 'aws', 'tableau', 'big_data']]
```

## How can you remove the numbers from the company name?

```python
data['Company Name'] = data['Company Name'].map(lambda x: x.split('\n')[0])
```

## How can you create some new features? (e.g. state column from the location column)

```python
data.columns
```

#### Add `state` column from the `location` column

```python
data['State'] = data['Location'].apply(lambda x: x.split(',')[-1].strip())

location_idx = data.columns.get_loc('Location')
data.insert(location_idx + 1, 'State_temp', data['State'])
data = data.drop('State', axis=1)
data = data.rename(columns={'State_temp': 'State'})
```

```python
data[['Location', 'State']]
```

#### Add `same_state` column from the `Location` column and `Headquarters` column

Look at this two columns, `Location` columns describes where you will work, in office in certain State in USA, `Headquarters` columns describes where is headquarter of company. If there is a same informations in these two columns it means you will work in headquarter of company, otherwise not, we will create new columns named `same_state`.

```python
data[['Location', 'Headquarters']]
```

```python
data['same_state'] = data.apply(lambda row: 1 if row['Location'] == row['Headquarters'] else 0, axis=1)

headquarters_idx = data.columns.get_loc('Headquarters')
data.insert(headquarters_idx + 1, 'same_state_temp', data['same_state'])
data = data.drop('same_state', axis=1)
data = data.rename(columns={'same_state_temp': 'same_state'})
```
